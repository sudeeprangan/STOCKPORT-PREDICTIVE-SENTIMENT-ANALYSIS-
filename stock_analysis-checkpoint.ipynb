# Dow Jones Sentiment Analysis

```python
import pandas as pd
import numpy as np
import re
from textblob import TextBlob
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
df1 = pd.read_csv("Dow_Jones_Industrial_Average_tweets.csv")
df1.tail()
```

```python
df2 = pd.read_csv("Dow_Jones_Industrial_Average_index.csv")
df2.head()
```

```python
df2.shape
```

```python
merge = df1.merge(df2, how="inner", on="Date")
```

```python
merge.tail()
```

```python
tweets = []
for row in range(0, len(merge.index)):
    tweets.append(' '.join(str(x) for x in merge.iloc[row, 2:27]))
tweets[0]
```

```python
clean_tweets = []
for i in range(0, len(tweets)):
    clean_tweets.append(re.sub("b[(')]", '', tweets[i]))
    clean_tweets[i] = re.sub('b[(")]', "", clean_tweets[i])
    clean_tweets[i] = re.sub("\'", "", clean_tweets[i])
clean_tweets[0]
```

```python
merge["Combined News"] = clean_tweets
merge.head()
```

```python
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity
```

```python
def getPolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
merge["Subjectivity"] = merge["Combined News"].apply(getSubjectivity)
merge["Polarity"] = merge["Combined News"].apply(getPolarity)
merge.head()
```

```python
def getSIA(text):
    sia = SentimentIntensityAnalyzer()
    sentiment = sia.polarity_scores(text)
    return sentiment
```

```python
compound = []
neg = []
pos = []
neu = []
SIA = 0
for i in range(0, len(merge["Combined News"])):
    SIA = getSIA(merge["Combined News"][i])
    compound.append(SIA["compound"])
    neg.append(SIA["neg"])
    pos.append(SIA["pos"])
    neu.append(SIA["neu"])
```

```python
merge["Compound"] = compound
merge["Positive"] = pos
merge["Negative"] = neg
merge["Neutral"] = neu
merge.head()
```

```python
df = merge[["Open", "High", "Low", "Volume", "Subjectivity", "Polarity", "Compound", "Negative", "Neutral", "Positive", "Label"]]
df.head()
```

```python
X = df
X = np.array(X.drop(["Label"], 1))
Y = np.array(df["Label"])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
model = LinearDiscriminantAnalysis().fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
print(accuracy_score(y_test, predictions))
print(classification_report(y_test, predictions))
print(confusion_matrix(y_test, predictions))
```

# Visualization

```python
def getSentiment(score):
    if score < 0:
        return "Negative"
    elif score == 0:
        return "Neutral"
    else:
        return "positive"
```

```python
merge["Sentiment"] = merge["Polarity"].apply(getSentiment)
```

```python
merge["Sentiment"].value_counts().plot(kind = "bar")
plt.xlabel("Sentiments")
plt.ylabel("Counts")
plt.title("Sentiment Analysis")
plt.show()
```

```python
plt.figure(figsize = (12.33, 4.5))
plt.title("Sentiment Sum over Time")
polarity = merge.groupby(["Date"]).sum()["Polarity"]
plt.plot(polarity.index, polarity)
plt.show()
```
